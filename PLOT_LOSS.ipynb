```r
rm(list=ls())
library(data.table)
library(ggplot2)
library(repr)
```

## 100bp

```r
log100bp=read.table("keratinocytes-0h.100bp_peak.profile.peaks.only.bpnet.0.log.batch",sep='\t',header=TRUE)
log100bp$Batch=seq(1,nrow(log100bp))
log100bp$scaled_count_loss=log100bp$logcount_predictions_loss*30
```

```r
options(repr.plot.width=10, repr.plot.height=5)
p1=ggplot(data=log100bp)+
geom_line(aes(alpha=0.5,x=Batch,y=loss,color="Training_Total"))+
geom_line(aes(alpha=0.5,x=Batch,y=logcount_predictions_loss,color="Training_LogCount"))+
geom_line(aes(alpha=0.5,x=Batch,y=profile_predictions_loss,color="Training_Profile"))+
geom_line(aes(alpha=0.5,x=Batch,y=scaled_count_loss,color="Training_LogCountx30"))+
xlab("Batch")+
ylab("Loss")+
ggtitle("Keratinocyte-0h,100bp\nBPNet w/300 Filters\nLog count loss x30")+
scale_color_manual(name="Loss",
      values = c(
           Training_LogCount='#1b9e77',
           Training_Profile='#d95f02',
           Training_Total='#7570b3',
           Training_LogCountx30='#000000'))+  
scale_y_continuous(trans = 'log2')+
theme_bw(20)
p1
```
